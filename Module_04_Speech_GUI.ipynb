```python
def downloadReport(pname, vf,jv,rep,shi,ppe,res):
    pdf = FPDF()
  
# Add a page
    pdf.add_page()
  
# set style and size of font 
# that you want in the pdf
    pdf.set_font("Arial", size = 15)
  
# create a cell
    pdf.cell(200, 10, txt = "PARKINSON DISEASE TEST REPORT (SPEECH)", 
         ln = 1, align = 'C')
  
# add another cell
    pdf.cell(200, 10, txt = "PatientName: "+pname,
         ln = 2, align = 'C')
    
    pdf.cell(200, 10, txt = "Voice Frequency: "+vf,
         ln = 2, align = 'C')

    pdf.cell(200, 10, txt = "JitterValue: "+jv,
         ln = 2, align = 'C')

    pdf.cell(200, 10, txt = "RAPValue: "+rep,
         ln = 2, align = 'C')

    pdf.cell(200, 10, txt = "ShimmerValue: "+shi,
         ln = 2, align = 'C')

    pdf.cell(200, 10, txt = "PPEValue: "+ppe,
         ln = 2, align = 'C')

    pdf.cell(200, 10, txt = "PATIENT'S RESULT   : "+res,
         ln = 2, align = 'C')
    
  
# save the pdf with name .pdf
    pdf.output("Speech REPORT "+pname+".pdf") 
```
