# Speech dataset

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
dfs= pd.read_csv('out5.csv')
```

```python
sns.heatmap(dfs.isnull(),yticklabels=False,cmap='viridis',cbar=False)
```

```python
for i in dfs:
    if i!='status' and i!='name':
        sns.catplot(x='status',y=i,kind='box',data=dfs)
```

```python
plt.figure(figsize=(15,15))
sns.heatmap(dfs.corr(), annot=True , cmap= "coolwarm")
```

```python
sns.set_style('whitegrid')
sns.countplot(x='status',data=dfs)
```

```python
n_patients = dfs.shape[0]   ##number of patients
n_features = dfs.shape[1]-1 ##number of features
n_parkinsons = dfs[dfs['status'] == 1].shape[0] ##persons with parkinson 
n_healthy = dfs[dfs['status'] == 0].shape[0] ##healthy persons
print ("total number of patients:", n_patients)
print ("number of features:", n_features)
print ("number of patients with parkinson:", n_parkinsons)
print ("number of patients without parkinsons:", n_healthy)
```

```python
#Histogram Plot of NHR distribution
dfs['NHR'].hist(figsize=(7,6), color='blue', alpha=0.7)
plt.xlabel('NHR')
plt.ylabel('No of Patients')
plt.title('NHR of distribution of patients')
```

# Gait dataset

```python
df2=pd.read_csv('g.csv')
```

```python
sns.heatmap(df2.isnull(),yticklabels=False,cmap='viridis',cbar=False)
```

```python
for i in df2:
    if i!='status' and i!='name':
        sns.catplot(x='class',y=i,kind='box',data=df2)
```

```python
plt.figure(figsize=(10,10))
sns.heatmap(df2.corr(), annot=True , cmap= "coolwarm")
```

```python
sns.set_style('whitegrid')
sns.countplot(x='class',data=df2)
```

```python
n_patients = df2.shape[0]   ##number of patients
n_features = df2.shape[1]-1 ##number of features
n_parkinsons = df2[df2['class'] == 1].shape[0] ##persons with parkinson 
n_healthy = df2[df2['class'] == 0].shape[0] ##healthy persons
print ("total number of patients:", n_patients)
print ("number of features:", n_features)
print ("number of patients with parkinson:", n_parkinsons)
print ("number of patients without parkinsons:", n_healthy)
```

```python
#Histogram Plot of Total VGRF Right distribution
df2['Total_force_right'].hist(figsize=(7,6), color='orange', alpha=0.7)
plt.xlabel('Total_force_right: VGRF')
plt.ylabel('No of Patients')
plt.title('Total VGRF Right by Patient counts')
```

```python
#Histogram Plot of Total VGRF left distribution
df2['Total_force_left'].hist(figsize=(7,6), color='black', alpha=0.7)
plt.xlabel('Total_force_left: VGRF')
plt.ylabel('No of Patients')
plt.title('Total VGRF Left by Patient counts')
```
