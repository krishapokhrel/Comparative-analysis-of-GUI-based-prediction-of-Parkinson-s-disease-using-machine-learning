```python
def downloadReport(pname, hs,hv,haccn,hj,res):
    pdf = FPDF()
  
# Add a page
    pdf.add_page()
  
# set style and size of font 
# that you want in the pdf
    pdf.set_font("Arial", size = 15)
  
# create a cell
    pdf.cell(200, 10, txt = "PARKINSON DISEASE TEST REPORT", 
         ln = 1, align = 'C')
  
# add another cell
    pdf.cell(200, 10, txt = "PatientName: "+pname,
         ln = 2, align = 'C')
    
    pdf.cell(200, 10, txt = "Handwriting speed: "+hs,
         ln = 2, align = 'C')

    pdf.cell(200, 10, txt = "Handwriting velocity: "+hv,
         ln = 2, align = 'C')

    pdf.cell(200, 10, txt = "Handwriting acceleration: "+haccn,
         ln = 2, align = 'C')

    pdf.cell(200, 10, txt = "Handwriting jerk: "+hj,
         ln = 2, align = 'C')


    pdf.cell(200, 10, txt = "Patient's result   : "+res,
         ln = 2, align = 'C')
    
  
# save the pdf with name .pdf
    pdf.output("Tremor Report "+pname+".pdf") 
```
