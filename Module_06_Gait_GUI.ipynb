```python
def svm():
    from sklearn import svm
    gnb = svm.SVC(kernel='linear')
    gnb=gnb.fit(X,np.ravel(y))

    # calculating accuracy-------------------------------------------------------------------
    from sklearn.metrics import accuracy_score
    y_pred=gnb.predict(X_test)
    #print(accuracy_score(y_test, y_pred))
    #print(accuracy_score(y_test, y_pred,normalize=False))
    # -----------------------------------------------------

    terms = [Gender.get(),VGRFLS.get(),VGRFRS.get()]
    for k in range(0,len(l6)):
        for z in terms:
            if(z==l6[k]):
                l7[k]=1

    inputtest = [l7]
    predict = gnb.predict(inputtest)
    predicted=predict[0]

    h='no'
    for a in range(0,len(patient)):
        if(predicted == a):
            h='yes'
            break

    if (h=='yes'):
        t4.delete("1.0", END)
        t4.insert(END, patient[a])
    else:
        t4.delete("1.0", END)
        t4.insert(END, "Not Found")
```

```python
def downloadReport(pname, page, gen,vgrfl,vgrfr,res):
    pdf = FPDF()
  
# Add a page
    pdf.add_page()
  
# set style and size of font 
# that you want in the pdf
    pdf.set_font("Arial", size = 15)
  
# create a cell
    pdf.cell(200, 10, txt = "PARKINSON DISEASE TEST REPORT (GAIT)", 
         ln = 1, align = 'C')
  
# add another cell
    pdf.cell(200, 10, txt = "PatientName: "+pname,
         ln = 2, align = 'C')
    
    pdf.cell(200, 10, txt = "Patient age: "+page,
         ln = 2, align = 'C')

    pdf.cell(200, 10, txt = "Gender Details: "+gen,
         ln = 2, align = 'C')

    pdf.cell(200, 10, txt = "VGRF Left Sensor Value: "+vgrfl,
         ln = 2, align = 'C')

    pdf.cell(200, 10, txt = "VGRF Left Sensor Value: "+vgrfr,
         ln = 2, align = 'C')

    pdf.cell(200, 10, txt = "Patient's result   : "+res,
         ln = 2, align = 'C')
    
  
# save the pdf with name .pdf
    pdf.output("Gait Report "+pname+".pdf") 
```
